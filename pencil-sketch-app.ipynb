 PROBLEM STATEMENT :-To develop a Python program that converts a given colour photograph into a realistic pencil sketch using image processing techniques in OpenCV.

```python
# ðŸ“Œ STEP 1: Install required library (OpenCV)
!pip install opencv-python

# ðŸ“Œ STEP 2: Import libraries
import cv2
from google.colab.patches import cv2_imshow  # for displaying images in Colab

# ðŸ“Œ STEP 3: Upload an image
from google.colab import files
uploaded = files.upload()

# Get the uploaded image filename
import io
import os

filename = list(uploaded.keys())[0]
print("Uploaded file:", filename)

# ðŸ“Œ STEP 4: Read the image
image = cv2.imread(filename)

# ðŸ“Œ STEP 5: Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ðŸ“Œ STEP 6: Invert the grayscale image
inverted_image = 255 - gray_image

# ðŸ“Œ STEP 7: Blur the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

# ðŸ“Œ STEP 8: Invert the blurred image
inverted_blur = 255 - blurred

# ðŸ“Œ STEP 9: Create the pencil sketch
pencil_sketch = cv2.divide(gray_image, inverted_blur, scale=256.0)

# ðŸ“Œ STEP 10: Show results
print("âœ… Original Image:")
cv2_imshow(image)

print("âœ… Pencil Sketch:")
cv2_imshow(pencil_sketch)

# ðŸ“Œ STEP 11: Save the sketch
output_file = "pencil_sketch.jpg"
cv2.imwrite(output_file, pencil_sketch)
print("Sketch saved as:", output_file)
```

This project successfully demonstrates how image processing techniques can be used to transform a colour photograph into a pencil sketch using Python and OpenCV. By applying steps such as grayscale conversion, image inversion, Gaussian blurring, and color dodge blending, we are able to generate a realistic sketch effect that closely resembles hand-drawn artwork.
